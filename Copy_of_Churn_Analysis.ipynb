```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Count occurrences
df_grouped = df.groupby(["SeniorCitizen", "Churn"]).size().unstack()

# Normalize to percentage
df_percent = df_grouped.div(df_grouped.sum(axis=1), axis=0) * 100

# Plot
fig, ax = plt.subplots(figsize=(5, 5))
df_percent.plot(kind="bar", stacked=True, colormap="coolwarm", ax=ax)

# Add percentage labels
for i, bars in enumerate(ax.containers):
    for bar in bars:
        height = bar.get_height()
        if height > 0:
            ax.text(
                bar.get_x() + bar.get_width() / 2,  # X position
                bar.get_y() + height / 2,  # Y position
                f"{height:.1f}%",  # Text
                ha="center", va="center", fontsize=10, color="white", fontweight="bold"
            )

# Labels & Title
ax.set_title("Churn by Senior Citizen Status (Stacked %)", fontsize=14)
ax.set_ylabel("Percentage")
ax.set_xlabel("Senior Citizen")
ax.legend(title="Churn", bbox_to_anchor = (0.9,0.9))

plt.show()
```

```python
# Define the columns to plot
cols = ['PhoneService', 'MultipleLines', 'InternetService',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies']

# Define the figure and axes
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 12))  # 3x3 grid
axes = axes.flatten()  # Flatten to easily iterate

# Loop through columns and create count plots
for i, col in enumerate(cols):
    sns.countplot(x=col, data=df, ax=axes[i],hue = 'Churn', palette="coolwarm")
    axes[i].set_title(f"Count of {col}")
    axes[i].set_xlabel("")  # Remove x-axis label for cleaner look
    axes[i].set_ylabel("Count")

# Adjust layout
plt.tight_layout()
plt.show()
```
